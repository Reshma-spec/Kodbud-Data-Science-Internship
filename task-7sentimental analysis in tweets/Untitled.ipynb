```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("test.csv")

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Sentiment function
def get_sentiment(tweet):
    score = analyzer.polarity_scores(str(tweet))['compound']

    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Create Sentiment column
df['Sentiment'] = df['tweets'].apply(get_sentiment)

# Show first 5 results
print(df[['tweets', 'Sentiment']].head())

# Count sentiments
print("\nSentiment Counts:")
print(df['Sentiment'].value_counts())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Sentiment')
plt.title('Sentiment Analysis of Tweets')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
df.to_csv("tweets_sentiment_results.csv", index=False)
print("Results saved successfully!")
```

```python
plt.figure(figsize=(6,6))
df['Sentiment'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title('Tweet Sentiment Distribution')
plt.ylabel('')
plt.show()
```

```python
sentiment_percentage = df['Sentiment'].value_counts(normalize=True) * 100

print("Sentiment Percentage:")
print(sentiment_percentage)
```

```python
summary = pd.DataFrame({
    'Count': df['Sentiment'].value_counts(),
    'Percentage': round(df['Sentiment'].value_counts(normalize=True) * 100, 2)
})

print(summary)
```

```python
positive_tweets = df[df['Sentiment'] == 'Positive']

print("Top Positive Tweets:")
print(positive_tweets['tweets'].head(10))
```
